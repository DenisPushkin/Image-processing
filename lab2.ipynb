```python
from matplotlib.pyplot import imread
from skimage.color import rgb2gray
import numpy as np
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
```

```python
alpha = 0.5
a, b = 1, 1
```

```python
img = np.int32(rgb2gray(imread(r'photo.jpg')) * 255)
plt.imshow(img, cmap='Greys_r')
print(img.shape)
print(img.max())
print(img.min())
print(img)
```

```python
fourier = np.fft.fft2(img)
#print(img_transformed.max())
#print(img_transformed.min())
#print(img_transformed)
phi = np.arctan2(fourier.imag, fourier.real)
fourier_transformed = np.abs(fourier)**alpha * np.exp(1j*phi)
img_transformed = np.fft.ifft2(fourier_transformed).real
max_val = np.max(img_transformed)
min_val = np.min(img_transformed)
img_transformed = np.floor((img_transformed - min_val)/(max_val-min_val)*255)
```
