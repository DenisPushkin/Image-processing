```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_image(image, title = 'Image', cmap_type = 'gray'):
    plt.imshow(image, cmap = cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()

def show_hist(image, title = 'Histogram'):
    plt.hist(image.ravel(), bins = 256)
    plt.title(title)
    plt.show()

def my_transformation(image, alpha = 0.5):
    fourier = np.fft.fft2(img)
    phi = np.arctan2(fourier.imag, fourier.real)
    fourier_transformed = np.abs(fourier)**alpha * np.exp(1j*phi)
    img_transformed = np.fft.ifft2(fourier_transformed).real
    return img_transformed

def truncate_extreme_values(image, p):
    image_values = image.ravel()
    image_values = np.sort(image_values)
    n = len(image_values)
    pos = np.int32(n * p)
    min_val = image_values[pos]
    max_val = image_values[n - pos]
    image_truncated = np.copy(image)
    image_truncated[image_truncated < min_val] = min_val
    image_truncated[image_truncated > max_val] = max_val
    return image_truncated
```

```python
alpha = 0.5
a, b = 1, 1
```

```python
img = plt.imread(r'C:\Users\111\Desktop\Image_processing\coursera\4.png')
print('image.shape =', img.shape)
show_image(img, title = 'Initial image')
show_hist(img, title = 'Initial image histogram')
```

```python
img_transformed = my_transformation(img, alpha = 0.6)
img_transformed_truncated = truncate_extreme_values(img_transformed, p = 0.05)
show_image(img_transformed_truncated, 'Transformed image, alpha = 0.6, p = 0.05')
show_hist(img_transformed_truncated, 'Transformed image hist, alpha = 0.6, p = 0.05')
```
